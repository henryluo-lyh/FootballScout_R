### Stats

'games_appearences'（比赛出场次数）：球员在比赛中出现的次数。

'games_lineups'（首发次数）：球员在比赛中首发的次数。

'games_minutes'（比赛分钟数）：球员在比赛中累计的出场分钟数。

'games_rating'（比赛评分）：球员在比赛中的表现得分。

'substitutes_in'（替补上场次数）：球员在比赛中被替补进场的次数。

'substitutes_out'（替补下场次数）：球员在比赛中被替换下场的次数。

'substitutes_bench'（替补席次数）：球员在比赛中坐在替补席上的次数。

'shots_total'（总射门次数）：球员在比赛中进行的总射门次数。

'shots_on'（射正次数）：球员在比赛中进行的射正次数。

'goals_total'（总进球数）：球员在比赛中打进的总进球数。

'goals_conceded'（失球数）：球员在比赛中被对方打进的进球数。

'goals_assists'（助攻数）：球员在比赛中的助攻次数。

'goals_saves'（扑救数）：球员在比赛中的扑救次数。

'passes_total'（总传球次数）：球员在比赛中进行的总传球次数。

'passes_key'（关键传球次数）：球员在比赛中进行的关键传球次数。

'passes_accuracy'（传球成功率）：球员在比赛中传球成功的比率。

'tackles_total'（总抢断次数）：球员在比赛中进行的总抢断次数。

'tackles_blocks'（封堵次数）：球员在比赛中进行的封堵次数。

'tackles_interceptions'（拦截次数）：球员在比赛中进行的拦截次数。

'duels_total'（总对抗次数）：球员在比赛中进行的总对抗次数。

'duels_won'（对抗成功次数）：球员在比赛中进行的对抗成功次数。

'dribbles_attempts'（带球次数）：球员在比赛中进行的带球次数。

'dribbles_success'（带球成功次数）：球员在比赛中进行的带球成功次数。

'dribbles_past'（过人次数）：球员在比赛比赛中进行的过人次数。

'fouls_drawn'（被犯规次数）：球员在比赛中被对方犯规的次数。

'fouls_committed'（犯规次数）：球员在比赛中犯规的次数。

'cards_yellow'（黄牌数）：球员在比赛中被出示黄牌的次数。

'cards_yellowred'（两黄变红次数）：球员在比赛中被出示两张黄牌变成红牌的次数。

'cards_red'（红牌数）：球员在比赛中被出示红牌的次数。

'penalty_won'（获得点球次数）：球员在比赛中获得点球的次数。

'penalty_commited'（犯规导致点球次数）：球员在比赛中犯规导致对方获得点球的次数。

'penalty_scored'（点球进球数）：球员在比赛中打进的点球数。

'penalty_missed'（点球未进数）：球员在比赛中未能打进的点球数。

'penalty_saved'（扑出点球次数）：球员在比赛中扑出的点球数。

'player_age'（球员年龄）：球员的年龄。

'player_height'（球员身高）：球员的身高。

'player_weight'（球员体重）：球员的体重。
